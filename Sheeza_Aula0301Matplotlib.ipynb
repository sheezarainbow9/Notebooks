```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
empresas = ['Apple', 'Google', 'Amazon',
            'Microsoft', 'Tencent',
            "McDonald's", 'Visa',
            'Facebook', 'Alibaba',
            'Louis Vuitton']
```

```python
bilhoes = [947.062, 819.573, 705.646,
           611.460, 214.023, 196.526,
           191.032, 186.421, 169.966,
           124.273]
```

```python
# já é possível plotar, só com as variáveis:
plt.plot(empresas, bilhoes)
```

```python
# embelezando o gráfico:
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
```

```python
# Redimensionando:
plt.rcParams['figure.figsize'] = (15, 8)

plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
plt.show()
```

```python
# Conhecendo o fontsize:
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.show()
```

```python
# Mudando a cor e legendando:
plt.plot(empresas, bilhoes, color = 'green')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.legend(['Bilhões de Dólares'], loc = 'right', fontsize=15)
plt.show()
```

```python
# Aumentar a espessura da linha (lw):
plt.plot(empresas, bilhoes, color = 'green', lw=8)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.legend(['Bilhões de Dólares'], loc = 'right', fontsize=15)
plt.show()
```

```python
# Adicionar marcador:
plt.plot(empresas, bilhoes, color = 'green', lw=2, marker = 'o')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.legend(['Bilhões de Dólares'], loc = 'right', fontsize=15)
plt.show()
```

```python
# Adicionando linhas de grade:
plt.plot(empresas, bilhoes, color = 'green', lw=2, marker = 'o')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.legend(['Bilhões de Dólares'], loc = 'right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
# Mudando o tipo de gráfico e colocando a legenda acima (high right):
plt.bar(empresas, bilhoes, color = 'green')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.legend(['Bilhões de Dólares'], loc = 'high right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
# Gráfico mais simples:
plt.bar(empresas, bilhoes)
```

```python
cores = ['red', 'yellow', 'green', 'blue', 'purple', 'orange', 'magenta', 'white', 'brown', 'black']
```

```python
plt.bar(empresas, bilhoes, color = cores, edgecolor='black')  # edgecolor = borda
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões de Dólares', fontsize=20)
plt.grid(True)
plt.show()
```

```python
plt.style.available  # importando estilos.
```

```python
plt.style.use('seaborn-dark-palette')
plt.bar(empresas, bilhoes, color = cores, edgecolor='black')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões de Dólares', fontsize=20)
plt.grid(True)
plt.show()
```

```python
# Tipo de gráfico barh:
plt.style.use('seaborn-dark-palette')
plt.barh(empresas, bilhoes, color = cores, edgecolor='black')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões de Dólares', fontsize=20)
plt.grid(True)
plt.show()
```

```python
plt.style.use('seaborn-dark-palette')
plt.scatter(empresas, bilhoes, color = cores, edgecolor='black')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões de Dólares', fontsize=20)
plt.grid(True)
plt.show()
```
