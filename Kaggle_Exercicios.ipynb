```python
# Crie uma variável 'df' com os valores 0, 1, 10, 100 e country, province, region_1, region_2:
df = reviews.loc[[0, 1, 10, 100], ['country', 'province', 'region_1', 'region_2']]
df
```

```python
# Vinhos italianos:
italian_wines = reviews['country'] == 'Italy'
reviews[italian_wines]
```

```python
# Selecionando vinhos de Austrália ou Nova Zelândia com pontos >= 95:
top_oceania_wines = (reviews['points'] >= 95) & (reviews['country'] == 'New Zealand') | (reviews['country'] == 'Australia')
reviews[top_oceania_wines]
```

```python
median_points = reviews.points.median()
median_points
```

```python
# Países presentes do dataset:
reviews.country.unique()
```

```python
# Total de vezes que os países aparecem no dataset:
reviews_per_country = reviews['country'].value_counts()
reviews_per_country
```

```python
# Coluna 'price' com a média subtraída:
centered_price = reviews.price - reviews.price.mean()
centered_price
```

```python
# Vinho mais bem pontuado e barato:
bargain_index = (reviews.points / reviews.price).idxmax()
bargain_wine = reviews.loc[bargain_index, 'title']
print(f'O vinho mais bem pontuado e com menor preço é o {bargain_wine}.')
```

```python
#Quantas vezes aparecem as palavras 'tropical' e 'fruity' na coluna 'description':

tropical = reviews.description.map(lambda d: 'tropical' in d).sum()
fruity = reviews.description.map(lambda d: 'fruity' in d).sum()
descriptor_counts = pd.Series([tropical, fruity], index=['tropical', 'fruity'])
descriptor_counts
```

```python
# Classificando vinhos por estrelas (1, 2, 3):
star_ratings = reviews.points.map(lambda p: 3 if (p >= 95) else 2 if (p>= 85) else 1)
star_ratings
```

```python
# Preços mínimo e máximo de cada vinho:
price_extremes = reviews.groupby(['variety']).price.agg([min, max])
price_extremes
```
