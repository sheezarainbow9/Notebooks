```python
# cópia do df1 para segurança dos dados.
df4 = df1.copy()
```

```python
# Deletando colunas específicas do DF:
df4.drop(columns=['data_notificacao', 'outros_sintomas', 'doencas_preexistentes',	'profissional_saude', 'classificacao_final', 'evolucao', 'municipio_residencia', 'bairro', 'distrito_sanitario', 'categoria_profissional', 'data_obito'], inplace=True)
df4
```

```python
# cópia do df1 para segurança dos dados.
df5 = df1.copy()
df5
```

```python
# Deletando valores NaN de uma coluna específica:
df5.dropna(subset = ['etnia'], inplace=True)
df5
```

```python
# Concatenando DF's:
df6 = pd.concat([df4, df5])
df6
```

```python
# cópia do df6 para segurança dos dados.
df7 = df6.copy()
```

```python
# Deletando colunas específicas do DF:
df7 = df7.drop(columns = ['etnia', 'distrito_sanitario', 'outros_sintomas', 'doencas_preexistentes', 'profissional_saude', 'categoria_profissional'])
df7
```

```python
# Agrupando por categoria para plotagem de gráfico simples:
df7.groupby(['idade']).idade.count().plot(figsize=(10, 5))
plt.xlabel('Idade', fontsize=15)
plt.ylabel('Quantidade', fontsize=15)
plt.title('Total de Idades', fontsize=20)
```

```python
# Agrupando por categoria para plotagem de gráfico do tipo em barras:
raca = df7.groupby(['raca']).raca.count().plot(kind='bar', width=0.3, figsize=(10, 5), color='purple')
plt.xlabel('Raça', fontsize=15)
plt.ylabel('Quantidade', fontsize=15)
plt.title('Total de Raça Declarado', fontsize=20)
```
